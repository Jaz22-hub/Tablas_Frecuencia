```python
# Importar bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set(style="whitegrid")

# Generar datos aleatorios
np.random.seed(42)  # Para reproducibilidad
data = np.random.uniform(low=0.0, high=100.0, size=500)
df = pd.DataFrame(data, columns=['Valores'])

# Crear tabla de frecuencias
frecuencia_absoluta = df['Valores'].value_counts(bins=10).sort_index()
frecuencia_relativa = frecuencia_absoluta / frecuencia_absoluta.sum()
frecuencia_acumulada = frecuencia_absoluta.cumsum()

# Crear un DataFrame para las tablas
tabla_frecuencia = pd.DataFrame({
    'Frecuencia Absoluta': frecuencia_absoluta,
    'Frecuencia Relativa': frecuencia_relativa,
    'Frecuencia Acumulada': frecuencia_acumulada
})

print(tabla_frecuencia)

# Gráfico de histograma para frecuencia absoluta
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.hist(df['Valores'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de Frecuencia Absoluta')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Absoluta')

# Gráfico de barras horizontales para frecuencia absoluta
plt.subplot(2, 2, 2)
frecuencia_absoluta.plot(kind='barh', color='salmon')
plt.title('Diagrama de Barras de Frecuencia Absoluta')
plt.xlabel('Frecuencia Absoluta')

# Gráfico de polígono de frecuencia relativa
plt.subplot(2, 2, 3)
plt.plot(frecuencia_relativa.index.mid, frecuencia_relativa, marker='o', color='orange')
plt.title('Polígono de Frecuencia Relativa')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Relativa')

# Gráfico de pastel para frecuencia relativa
plt.subplot(2, 2, 4)
plt.pie(frecuencia_relativa, labels=frecuencia_relativa.index.astype(str), autopct='%1.1f%%', startangle=140)
plt.title('Diagrama de Pastel de Frecuencia Relativa')

plt.tight_layout()
plt.show()

# Gráfico de ojiva para frecuencia acumulada
plt.figure(figsize=(8, 5))
plt.plot(frecuencia_acumulada.index.mid, frecuencia_acumulada, marker='o', color='purple')
plt.title('Ojiva de Frecuencia Acumulada')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Acumulada')
plt.grid()
plt.show()
```
