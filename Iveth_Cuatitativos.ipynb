```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generar 500 datos enteros aleatorios (por ejemplo, entre 1 y 20)
np.random.seed(42)  # Semilla para reproducibilidad
datos = np.random.randint(1, 21, size=500)

# 2. Crear tabla de frecuencias
valores, conteo = np.unique(datos, return_counts=True)

tabla_frecuencia = pd.DataFrame({
    "Valor": valores,
    "Frecuencia Absoluta": conteo,
})
tabla_frecuencia["Frecuencia Relativa"] = tabla_frecuencia["Frecuencia Absoluta"] / len(datos)
tabla_frecuencia["Frecuencia Acumulada"] = tabla_frecuencia["Frecuencia Absoluta"].cumsum()

# Mostrar la tabla
print("Tabla de frecuencias:")
print(tabla_frecuencia)

# Graficar la frecuencia absoluta usando un histograma
plt.figure(figsize=(12, 6))
plt.hist(datos, bins=np.arange(0.5, 21.5, 1), color='lightblue', edgecolor='black')
plt.title('Histograma de Frecuencia Absoluta')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(range(1, 21))
plt.grid(axis='y')
plt.show()

# Graficar la frecuencia absoluta usando un diagrama de barras horizontales
plt.figure(figsize=(12, 6))
tabla_frecuencia.plot(kind='barh', x='Valor', y='Frecuencia Absoluta', color='lightgreen', edgecolor='black')
plt.title('Diagrama de Barras de Frecuencia Absoluta')
plt.xlabel('Frecuencia Absoluta')
plt.ylabel('Valores')
plt.grid(axis='x')
plt.show()

# Graficar la frecuencia relativa usando un polígono de frecuencia
plt.figure(figsize=(12, 6))
plt.plot(tabla_frecuencia['Valor'], tabla_frecuencia['Frecuencia Relativa'], marker='o', color='orange')
plt.title('Polígono de Frecuencia Relativa')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Relativa')
plt.grid()
plt.show()

# Graficar la frecuencia relativa usando un diagrama de pastel
plt.figure(figsize=(8, 8))
plt.pie(tabla_frecuencia['Frecuencia Relativa'], labels=tabla_frecuencia['Valor'], autopct='%1.1f%%', startangle=140)
plt.title('Diagrama de Pastel de Frecuencia Relativa')
plt.axis('equal')  # Para que el pastel sea circular
plt.show()

# Graficar la frecuencia acumulada usando la ojiva
plt.figure(figsize=(12, 6))
plt.plot(tabla_frecuencia['Valor'], tabla_frecuencia['Frecuencia Acumulada'], marker='o', color='purple')
plt.title('Ojiva de Frecuencia Acumulada')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Acumulada')
plt.grid()
plt.show()
```
